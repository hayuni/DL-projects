```python
import requests
from langdetect import detect

url = "https://defined.ai/dataset/healthcare-prompt-and-response-data"

response = requests.get(url)

if response.status_code == 200:
    content = response.content.decode("utf-8")
    lines = content.split("\n")

    # Filter English Content
    english = []
    for line in lines:
        try:
            if detect(line) == "en":
                english.append(line)
        except:
            pass

    # Join the English lines
    english_content = "\n".join(english)

    print(english_content)
else:
    print("Failed to access  URL:", response.status_code)
```
